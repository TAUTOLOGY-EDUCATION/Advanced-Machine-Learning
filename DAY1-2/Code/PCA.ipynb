```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA

# Data Generation
X, y = make_classification(n_samples=100, n_features=10, n_classes=2, random_state=42)

# Display Data Before PCA
data_before_pca = pd.DataFrame(X, columns=[f'Feature {i+1}' for i in range(X.shape[1])])
print("Data Before PCA:")
print(data_before_pca)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plotting after PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=80)
plt.title('PCA Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class')
plt.grid(True)
plt.show()
```
