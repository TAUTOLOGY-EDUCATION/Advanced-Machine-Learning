```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Data generation
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Plot before clustering
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Data Before AHC Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# AHC clustering
threshold = 15
ahc = AgglomerativeClustering(distance_threshold=threshold, n_clusters=None)
y_ahc = ahc.fit_predict(X)

# Compute linkage matrix
Z = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(12, 8))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Point Index')
plt.ylabel('Distance')
dendrogram(Z)
plt.axhline(y=threshold, color='r', linestyle='--', linewidth=2)
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_ahc, s=50, cmap='viridis')
plt.title('Data Points After Agglomerative Hierarchical Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
